```python
import pandas
import itertools
import pandas as pd
import tabulate
from tabulate import tabulate
# check duplicated rows based on a specific column
def locate_duplicate_rows(path,sep,check_col_name):
    table = pd.read_csv(path,sep)
    nb_duplicate_rows = table[check_col_name].duplicated().sum()
    print("Number of duplicated rows:" ,nb_duplicate_rows)
    print("---------------------------------------------------------------------------------")
    table_dup_rows = table.loc[table[check_col_name].duplicated(keep = False),:] # keep = 'last'/ keep = False
    dup_row_index = table_dup_rows.index + 1 #location of rows
    dup_col_index = table.columns.get_loc(check_col_name) + 1 #location of the specific column
    dup_row_num = []
    for i in dup_row_index : 
        for j in dup_row_index : 
            if table.iat[i-1,dup_col_index-1] == table.iat[j-1,dup_col_index-1] and i != j:
                print("row index:",i,"col index:",dup_col_index)
                print("duplicated value is:",table.iat[i-1,dup_col_index-1])
                print("duplicated value is located in row", j)
                print("---------------------------------------------------------------------------------")
                dup_row_num.append(j)
    print("the list of duplicated row index :", dup_row_num)
         

if __name__ == '__main__':
	# give filepath, seperator, the column name you want to check
	file = r"C:/Users/hanww/Documents/Python/check_duplicated_rows/CZ_2019_07-15-2019_athena.csv"
	seperator = ","
	col = "job_url"
	print(locate_duplicate_rows(file,seperator,col))



```
